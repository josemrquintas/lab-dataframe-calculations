Next, extract and import `Orders` dataset into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
# your code here
with zipfile.ZipFile("Orders.zip","r") as zip_ref:
    zip_ref.extractall("targetdir")

orders = pd.read_csv('targetdir/Orders.csv')
orders.head()

len(set(orders['CustomerID']))
```

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?


#### Sub Problem 2: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here

#Starting by aggregating customers by amount_spent

customer_amount = orders.groupby('CustomerID').agg({'amount_spent': np.sum })

customer_amount

# Labelling Customers by amount spent. 
sorted_orders = (pd.merge(orders, customer_amount))

[0, .75, .95, 1.]
bins = ['Standard', 'Preferred','VIP']

sorted_orders['Type of Customer'] = pd.qcut(sorted_orders['amount_spent'], [0, .75, .95, 1.], labels = bins)

sorted_orders.shape

```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
vip_customers = sorted_orders[sorted_orders['Type of Customer'] == 'VIP']
vip_customers.head()
```

```python
vip_customers.groupby('Country').agg({'CustomerID':'count'})
```

```python
(vip_customers.groupby('Country').agg({'CustomerID':'count'})).idxmax()
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
top_customers = sorted_orders[(sorted_orders['Type of Customer'] == 'VIP') | (sorted_orders['Type of Customer'] == 'Preferred')]
```

```python
top_customers.groupby('Country').agg({'CustomerID':'count'})
```

```python
(top_customers.groupby('Country').agg({'CustomerID':'count'})).idxmax()
```
